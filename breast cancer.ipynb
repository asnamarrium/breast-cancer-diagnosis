# Importing Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# A. Data Preparation

```python
# Loading dataset and inspecting it
df = pd.read_csv('breast_cancer.csv')
df = df.drop(columns=["id"])
df = df.drop(columns=["Unnamed: 32"])
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Handling missing values
df.isnull().sum().sum()
```

```python
df.isnull().sum()
```

```python
# Encoding categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["diagnosis"] = le.fit_transform(df["diagnosis"])  
print(df['diagnosis'].value_counts())
```

```python
df.head()
```

```python
X = df.drop(columns=["diagnosis"])
y = df["diagnosis"]
```

```python
# Scaling numerical features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Split into train/test sets (80/20)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=42,stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

# B. Feature Selection

```python
# Feature Selection (Embedded Method - Random Forest)
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
```

```python
rf.fit(X_train,y_train)
```

```python
importances = rf.feature_importances_
```

```python
feature_importances = pd.DataFrame({
    "Feature": X.columns,   
    "Importance": importances
}).sort_values(by="Importance", ascending=False)
```

```python
k = 10
selected_features = feature_importances.head(k)['Feature'].tolist()
print(selected_features)
```

```python
X_selected = X[selected_features]
X_selected = X[selected_features]
```

```python
X_train_selected, X_test_selected, y_train_selected, y_test_selected = train_test_split(
    X_selected, y, test_size=0.2, random_state=42,stratify=y)
```

```python
X_train_selected.shape
```

```python
X_test_selected.shape
```

# C. Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
# Part 1: Without Feature Selection
# Logistic Regression
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
# Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train,y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
# Part 2: With Feature Selection
# Logistic Regression (selected features)
lr_fs = LogisticRegression(max_iter=1000, random_state=42)
lr_fs.fit(X_train_selected, y_train_selected)
```

```python
y_pred_lr_fs = lr_fs.predict(X_test_selected)
```

```python
# Random Forest (selected features)
rf_fs = RandomForestClassifier(random_state=42)
rf_fs.fit(X_train_selected, y_train_selected)
```

```python
y_pred_rf_fs = rf_fs.predict(X_test_selected)
```

# D. Model Evaluation and Comparison

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Logistic Regression
y_pred_lr = lr.predict(X_test)
acc_lr = accuracy_score(y_test, y_pred_lr)
prec_lr = precision_score(y_test, y_pred_lr)
rec_lr = recall_score(y_test, y_pred_lr)
f1_lr = f1_score(y_test, y_pred_lr)

# Random Forest
y_pred_rf = rf.predict(X_test)
acc_rf = accuracy_score(y_test, y_pred_rf)
prec_rf = precision_score(y_test, y_pred_rf)
rec_rf = recall_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)


# Logistic Regression with FS
y_pred_lr_fs = lr_fs.predict(X_test_selected)
acc_lr_fs = accuracy_score(y_test, y_pred_lr_fs)
prec_lr_fs = precision_score(y_test, y_pred_lr_fs)
rec_lr_fs = recall_score(y_test, y_pred_lr_fs)
f1_lr_fs = f1_score(y_test, y_pred_lr_fs)

# Random Forest with FS
y_pred_rf_fs = rf_fs.predict(X_test_selected)
acc_rf_fs = accuracy_score(y_test, y_pred_rf_fs)
prec_rf_fs = precision_score(y_test, y_pred_rf_fs)
rec_rf_fs = recall_score(y_test, y_pred_rf_fs)
f1_rf_fs = f1_score(y_test, y_pred_rf_fs)
```

```python

# Create Comparison Table

results = pd.DataFrame({
    "Model": ["LogReg (No FS)", "RandomForest (No FS)", "LogReg (FS)", "RandomForest (FS)"],
    "Accuracy": [acc_lr, acc_rf, acc_lr_fs, acc_rf_fs],
    "Precision": [prec_lr, prec_rf, prec_lr_fs, prec_rf_fs],
    "Recall": [rec_lr, rec_rf, rec_lr_fs, rec_rf_fs],
    "F1-Score": [f1_lr, f1_rf, f1_lr_fs, f1_rf_fs]
})

print("Model Comparison Table")
print(results)


# Confusion Matrix
print("Confusion Matrix - Logistic Regression (FS)", confusion_matrix(y_test, y_pred_lr))
print("Confusion Matrix - Random Forest (FS)", confusion_matrix(y_test, y_pred_rf))
print("Confusion Matrix - Logistic Regression (FS)", confusion_matrix(y_test, y_pred_lr_fs))
print("Confusion Matrix - Random Forest (FS)", confusion_matrix(y_test, y_pred_rf_fs))
```
